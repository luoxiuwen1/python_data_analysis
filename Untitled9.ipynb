# 清洗数据
invoicedata，customerid 数据类型进行转换

quantity 和 unitprice 为负数的观察值删除

description缺失值删除

相同国家用不同方式的替换成

```python
clean_data=original_data.copy()
clean_data
```

```python
clean_data["InvoiceDate"]=pd.to_datetime(clean_data["InvoiceDate"])
clean_data["CustomerID"]=clean_data["CustomerID"].astype(str)
```

```python
clean_data["CustomerID"]=clean_data["CustomerID"].str.slice(0,-2)
clean_data["CustomerID"]
```

```python
clean_data.dropna(subset=["Description"],inplace=True)
clean_data["Description"].isnull().sum()
```

```python
clean_data["Country"].replace({"USA":"United States","UK":"United Kingdom","U.K.":"United Kingdom"})
```

```python
clean_data=clean_data[clean_data["Quantity"]>=0]
```

```python
clean_data.to_csv("D:\参考文献\e_commerce.csv",index=False)
```
